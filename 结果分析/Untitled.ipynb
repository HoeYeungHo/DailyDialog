```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
from matplotlib.font_manager import FontProperties

plt.rcParams['font.family'] = 'Times New Roman', 'SimSun'
plt.rcParams['mathtext.fontset'] = 'cm'
font = FontProperties(fname='SourceHanSerifCN-Heavy.ttf')
```

```python
# 设置备选颜色
color1 = "#e86254"
color2  = "#ef8a46"
color3  = "#f8aa59"
color4 = "#ffd06e"
color5  = "#fee6b5"
color6  = "#abdce0"
color7 = "#73bcd5"
color8  = "#528fac"
color9 = "#386795"
color10 = "#1f466f"

colorMain = [color1, color4, color7, color9]
colorALL = eval("[{}]".format(",".join(["color{}".format(i) for i in range(1, 11)])))
```

```python
# 加载数据
data = pd.read_excel(r"data/结果记录表.xlsx")

for i in range(len(data)):
    OpenString = "data{} = pd.read_csv('data/{}.csv')".format(i + 1, data["文件名"][i])
    exec(OpenString)
    print(OpenString)
```

```python
def saveFig(title):
    plt.savefig("SVG/{}.svg".format(title), format="svg", bbox_inches='tight')
    plt.savefig("PDF/{}.pdf".format(title), format="pdf", bbox_inches='tight')
```

## 基准

```python
plt.figure(figsize=(12, 4), dpi = 300)
# 情感分类
plt.plot(data1.batchALL, data1.EmotionACC, label = "情感分类准确率", color = colorMain[0])
plt.hlines(data["EmotionACC"][0], data1.batchALL.min(), data1.batchALL.max(), alpha = 0.5, linestyle = ":", color = colorMain[0])
# 行动
plt.plot(data1.batchALL, data1.ActACC, label = "行动分类准确率", color = colorMain[1])
plt.hlines(data["ActACC"][0], data1.batchALL.min(), data1.batchALL.max(), alpha = 0.5, linestyle = ":", color = colorMain[1])
# 总
plt.plot(data1.batchALL, data1.ALLACC, label = "总分类准确率", color = colorMain[3])
plt.hlines(data["ALLACC"][0], data1.batchALL.min(), data1.batchALL.max(), alpha = 0.5, linestyle = ":", color = colorMain[3])
# 散点
plt.scatter(data1[data1.EmotionACC > 0.65].batch.iloc[0], data1[data1.EmotionACC > 0.65].EmotionACC.iloc[0], color = colorALL[1], alpha = 0.5)
plt.scatter(data1.batchALL.iloc[data1.EmotionACC.argmax()], data1.EmotionACC.iloc[data1.EmotionACC.argmax()], color = colorALL[2], alpha = 0.5)

legend = plt.legend()

for text in legend.get_texts():
    if text.get_text() == '情感分类准确率':
        text.set_fontproperties(font)

saveFig("基准")
```

## 平衡因子

```python
width = 0.5

plt.figure(figsize=(8, 5), dpi = 300)

labels = [r"$\alpha={}$".format(i) for i in [0, 0.5, 1, 5, "\infty"]]

for i, y in enumerate((data.EmotionACC[4], data.EmotionACC[5], data.EmotionACC[0], data.EmotionACC[6], 0)):
    plt.bar(width * i, y, width=width, color = colorALL[i], label = labels[i])
    if y != 0: plt.text(width * i, y + 0.003, y, ha='center')
    else: plt.text(width * i, 0.302, y, ha='center')
for i, y in enumerate((0, data.ActACC[5], data.ActACC[0], data.ActACC[6], data.ActACC[7])):
    plt.bar(width * (5 + i), y, width=width, color = colorALL[i])
    if y != 0: plt.text(width * (5 + i), y + 0.003, y, ha='center')
    else: plt.text(width * (5 + i), 0.302, y, ha='center')

plt.xticks([width * 2, width * 7], ["情感分类准确率", "行动分类准确率"])
plt.ylim(0.3, 0.649)
plt.legend(loc = "upper right")

saveFig("平衡因子")
```
