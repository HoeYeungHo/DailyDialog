```python
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# 设置字体族，中文为SimSun，英文为Times New Roman
plt.rcParams['font.family'] = 'Times New Roman', 'SimSun'
plt.rcParams['mathtext.fontset'] = 'cm'  # 设置数学公式字体为stix
```

```python
def saveFig(title):
    plt.savefig("SVG/{}.svg".format(title), format="svg", bbox_inches='tight')
    plt.savefig("PDF/{}.pdf".format(title), format="pdf", bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4), dpi = 180)

# 绘制第一个箱线图和第一个y轴
ax1.boxplot(sentenceNum)
ax1.set_ylabel('每个对话中对话的轮数')

# 创建第二个y轴
ax2 = ax1.twinx()
ax2.boxplot(wordNum, positions=[1.5])
ax2.set_ylabel('每个句子中单词个数')

# 隐藏第一个y轴的ticks
ax1.tick_params(axis='y')
# 设置x轴ticks位置
ax1.set_xticks([1, 1.5])
ax1.set_xticklabels(['每个对话中对话的轮数', '每个句子中单词个数'])

saveFig("数据分析")
```
